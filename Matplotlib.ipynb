## Customizing plots

- Add legend `legend()`
- Adjust axes `axis([xmin, ymin, xmax, ymax])` 
- set axes labels `xlabel()` , `ylabel()` 
- Save figure `savefig()`

Lets recreate our previous plot and customize it one new command at a time.

```python
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x,y1,"bo-",linewidth=2, markersize=4);
plt.plot(x,y2,"gs-",linewidth=2, markersize=4);
```

Set axes labels `xlabel()` , `ylabel()` 
_**NOTE**_: the labels below are LateX formatted. Remove the `$` in the label function to see the subtle difference it makes

```python
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x,y1,"bo-",linewidth=2, markersize=4);
plt.plot(x,y2,"gs-",linewidth=2, markersize=4);
plt.xlabel("$X$");
plt.ylabel("$Y$");
```

```python
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x,y1,"bo-",linewidth=2, markersize=4);
plt.plot(x,y2,"gs-",linewidth=2, markersize=4);
plt.xlabel("$X$");
plt.ylabel("$Y$");
plt.axis([-0.5, 10.5, -5, 105]);
```

Add keyword argument `label` to the plot function and add legend to the plot with `plt.legend()`

```python
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5
plt.plot(x,y1,"bo-",linewidth=2, markersize=4, label="First");
plt.plot(x,y2,"gs-",linewidth=2, markersize=4, label="Second");
plt.xlabel("$X$");
plt.ylabel("$Y$");
plt.axis([-0.5, 10.5, -5, 105]);
plt.legend(loc="upper left")
```

To save the figure, `plt.savefig()` with plot name (along with path if you don't want the plot in your working director) as argument.

## Plotting Using Logarithmic Axes

- Plot x-axis _only_ in log scale i.e. y-axis in original scale `semilogx()` 
- Plot y-axis _only_ in log scale `semilogy()` 
- both axes in log scale `loglog()`

```python
x = np.linspace(0,10,20)
y1 = x**2.0
y2 = x**1.5
plt.loglog(x,y1,"bo-",linewidth=2, markersize=4, label="First");
plt.loglog(x,y2,"gs-",linewidth=2, markersize=4, label="Second");
plt.xlabel("$X$");
plt.ylabel("$Y$");
#plt.axis([-0.5, 10.5, -5, 105]);
plt.legend(loc="upper left")
```

```python
#x = np.logspace(-1,1,40)
x = np.logspace(np.log10(0.1),np.log10(10),40)
y1 = x**2.0
y2 = x**1.5
plt.loglog(x,y1,"bo-",linewidth=2, markersize=4, label="First");
plt.loglog(x,y2,"gs-",linewidth=2, markersize=4, label="Second");
plt.xlabel("$X$");
plt.ylabel("$Y$");
#plt.axis([-0.5, 10.5, -5, 105]);
plt.legend(loc="upper left")
```

## Histograms



```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.normal(size=1000)
plt.hist(x)
```

Lets customize the histogram !
By default `hist` uses 10 evenly spaced bins and tries to optimize both bin width and location.

In the plot below, instead of having the number of observations that fall within each bin (see plot above), we will have the proportion of observations that fall in each bin on y-axis 

```python
x = np.random.normal(size=1000)
plt.hist(x, normed=True)
```

To provide the location of bins,
below we are going to specify 20 bins between -5 and 5

```python
x = np.random.normal(size=1000)
plt.hist(x, normed=True, bins=np.linspace(-5,5,21))
```

`subplot()` enables us to have several subplots within each figure. It takes three arguments : no. of rows, no. of columns, plot no. e.g. 2 rows and 3 cols will give 6 subplots (numbering starts from 1 incrementing across rows)
_**Note**_ : this is similar to the `par()` function in R

```python
#x = np.random.gamma(2,3, size= 100000)
x = np.random.gamma(shape=2, scale=3, size= 100000)
plt.hist(x)
```

```python
plt.hist(x, bins=30)
```

```python
plt.hist(x, bins=30, normed=True)
```

```python
plt.hist(x, bins=30, cumulative=True)
```

```python
plt.hist(x, bins=30, cumulative=True, normed=True, histtype="step")
```

Now lets plot all these histogams together !

```python
plt.figure()
plt.subplot(221) # 2 rows x 2 cols fig 1
plt.hist(x, bins=30)
plt.subplot(222) # 2 rows x 2 cols fig 2
plt.hist(x, bins=30, normed=True)
plt.subplot(223) # 2 rows x 2 cols fig 3
plt.hist(x, bins=30, cumulative=30)
plt.subplot(224) # 2 rows x 2 cols fig 4
plt.hist(x, bins=30, normed=True, cumulative=True, histtype="step")
```
